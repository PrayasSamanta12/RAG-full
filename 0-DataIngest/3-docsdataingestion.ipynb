### Loading Docx in the Langchain ingestion
Libraries using to load docs - pyhton-docx,docx2txt

```python
import os
from langchain_community.document_loaders import Docx2txtLoader,UnstructuredFileLoader
```

```python
import os
from docx import Document as DocxDocument

# Define file path
output_dir = "data/docfiles"
output_file = os.path.join(output_dir, "SampleDocFile.docx")

# ✅ Ensure the directory exists
os.makedirs(output_dir, exist_ok=True)

# ✅ Create the document
doc = DocxDocument()
doc.add_heading('Sample Document', level=0)
doc.add_paragraph('This is a sample paragraph in the document.')
doc.add_paragraph('This document is created for testing purposes.')
doc.add_paragraph('LangChain document ingestion with Docx2txtLoader.LangChain document ingestion with Docx2txtLoader.\nLangChain document ingestion with Docx2txtLoader.\nLangChain document ingestion with Docx2txtLoader.')
doc.add_paragraph('End of the sample document.')
# ✅ Save it to the desired location
doc.save(output_file)

print(f"Document successfully created at: {output_file}")
```

```python
document=Docx2txtLoader("data/docfiles/SampleDocFile.docx").load()
print(f"Total documents loded: {len(document)}")
for doc in document:
    print(f"Document Content: {doc.page_content}")
    print(f"Metadata: {doc.metadata}")
```

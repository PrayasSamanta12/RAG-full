```python
from langchain_huggingface import HuggingFaceEmbeddings
import numpy as np
```

```python
sentences = [
    "The cat sat on the mat",
    "The dog sat on the mat",
    "The dog played in the yard",
    "I love programming in Python",
    "Python is my favorite programming language"
]
```

```python
#Function of cosine similarity
import numpy as np
def cosine_similarity(vec1,vec2):
    """
    Calculate cosine similarity between two vectors
    - Result close to 1: Similar meanings
    - Result close to 0: Unrelated meanings 
    """
    dot_product=np.dot(vec1,vec2)
    norm_a=np.linalg.norm(vec1)
    norm_b=np.linalg.norm(vec2)
    return dot_product/(norm_a*norm_b)
```

```python

import numpy as np
embeddings_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

for i in range(len(sentences)):
    for j in range(i, len(sentences)):
        vec1 = embeddings_model.embed_query(sentences[i])
        vec2 = embeddings_model.embed_query(sentences[j])
        similarity = cosine_similarity(np.array(vec1), np.array(vec2))
        print(f"Cosine similarity between:\n'{sentences[i]}'\nand\n'{sentences[j]}'\n is: {similarity}\n")


```
